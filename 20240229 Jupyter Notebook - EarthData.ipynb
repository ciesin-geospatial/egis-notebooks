## Accessing Functions and Directories of the ArcGIS Python Package

We'll first look at how to access the inforamtion in the ArcGIS Python packages using an example dataset from the NASA EarthData site. 

We'll use the data on monthly precipitation from the NASA NEX-GDDP-CMIP6 downscaled product for the Canadian Earth System Model version 5 (CanESM5) climate model and the Shared Socioeconomic Pathyway (SSP126) scenario. This is a data product that has taken future projections (2015-2100) for the monthly precipitation outputs of the CanESM5 model and scaled them down to a finer scale, since most climate outputs are on a rather course scale.

In this section, we import the necessary packages, specify the URL of the data product, and observe the directories and functions available to us.

```python
# Import the necessary packages

import arcgis

# Specify the url where the data is located

esm_url = r'https://gis.earthdata.nasa.gov/UAT/rest/services/NEX-GDDP-CMIP6/pr_monthly_CanESM5_ssp126/ImageServer'

# Provide the url to the ArcGIS package and access the directories and functions of the data product

pr_monthly = arcgis.raster.ImageryLayer(esm_url)

print(dir(pr_monthly))
```

As we can see, there are quite a few options available to us. Below, we'll take a look at how to get a closer look at the data, plot simple and customized histograms of the data, view simple statistics, and how to acquire values for individual locations and rasters.

## Accessing the Properties of the Dataset

To get a better sense of the dataset, let's look at the properties of it using the ArcGIS package and the "properties" function we saw in the last section.

```python
# Display the properties of the dataset

display(pr_monthly.properties)
```

We can see useful information from the properties listed, such as the full name, extent of the of dataset, its spatial reference system, how the pixels are stored, some basic statistics, whether its multidimensional, along with other info.

## Accessing the Statistics of the Dataset

As we saw in the last section, some basic statistics were provided in the properties. However, the ArcGIS package also allows us to access this information directly with the statistics function.

```python
# Display primary statistics of the dataset

display(pr_monthly.statistics())
```

From here, we can see some basic statistics about the monthly precipitation values in the dataset. The minimum precipitation for any month was 0.00 inches, the maximum was 92.96 inches, the mean was 2.45 inches, and the standard deviation was 3.27 inches.

## Accessing Histogram Information via the ArcGIS Package

Although the basic statistics information in the last section are useful, there are additional methods that allow us to get an even closer look at the data that are native to the ArcGIS package. In this section, we'll leverage the "get_histograms" and the "plot_histograms" function.

The "get_histograms" function shows the number of bins in a sample histogram and the number of observations in each of those bins, along with the minimum and maximum.

```python
# Display the histogram information

display(pr_monthly.get_histograms())
```

The "plot_histograms" function takes the information seen above and plots it into a histogram.

```python
# Display the histogram

pr_monthly.plot_histograms()
```

From the above histogram, we can see that the precipitation values are positively skewed with the majority of of them being less than about 15 inches per month. Additionally, we can see the first standard deviation above the mean is about 6 inches per month.

## Getting Precipitation Values for Specific Locations and Times

One of the most useful features is being able to query specific locations and times of the rasters (in this case, precipitation).

In the cell below we first acquire all of the time values available from each slice. The time values provided for each slice are provided in 'epoch' time, which is a unit of measure that provides a time in Unix code from a particular start date. For this tutorial, you do not need to worry about the specifics.

Next, we convert the 'epoch' times into GMT times.

Finally, we store those values in a pandas dataframe and display them for visualization.

```python
# Import necessary packages

import datetime
import pandas as pd

# Get a list of the slices available from the raster

slice_list = pr_monthly.slices()['slices']

# Initialize a variable called "times" that will store the time information of each slice

times = {}

# Create a for-loop that cycles through each of the slices, gets the 'epoch' time for the slice, and convert it to GMT

for slices in slice_list:
    ID = slices['sliceId'] # Get the of the respective slice from the slice_list
    multi_def = slices['multidimensionalDefinition'][0]
    epoch_time = multi_def['values'][0] # Get the value of the 'epoch' time for that slice
    GMT_time = datetime.datetime.fromtimestamp(epoch_time/1000) # Convert the 'epoch' time to GMT
    times[ID] = [epoch_time,GMT_time] # Store the 'epoch' and GMT times

# Convert the times variable into a dataframe

times_df = pd.DataFrame.from_dict(times,orient='index',columns=['Epoch Time','GMT Time'])

# Display the times dataframe

display(times_df)
```

We can see that each slice is representative of a particular month for each year going from January 2015 through December 2100.

Next, we'll isolate the times that are associated for the month of July. We'll extract them for each year from 2015 to 2100, store them in a new dataframe, and visualize them.

```python
# Keep only the time slices that are associated with July

filtered_times_df = times_df.iloc[6::12]
filtered_times_df
```

From the visual inspection of the filtered times dataframe we can see that we have successfully isolated the approapriate times.

Now that we have our time points, let's pick our spatial points using latitude and longitude.

Below, we specify a range of latitudes between 24 and 48 degrees North and a range of longitudes between 66 and 124 degrees West. This area encompasses the continental US. For this tutorial, we specify 7 points of latitude and 10 points of longitude evenly spaced between the range we specified. However, any number of latitudes and longitudes can be provided by simply editting the numbers below. Using these ranges, we then create a grid of latitude and longitude pairs for each of those points.

Once we have our latitude and longitude pairs, along with our time points, we query the NASA server to provide a value of precipitation from the CanESM5 model using the "identify" function and store these in an array. Finally, we move these to the precipitation dataframe and visualize it.

**Note:** Because the "identify" function must query the server one point at a time, it can take several minutes to acquire all of the values. We have kept the grid relatively coarse to limit the time required for this part of the tutorial. If you choose to use more point, it will add to the time required. For convenience, a time estimate is provided in the output that updates after each iteration while running the next cell.

```python
import time
from IPython.display import display, clear_output
import numpy as np

lats = np.linspace(24,48,7)
lons = np.linspace(-124,-66,10)

lat_grid,lon_grid = np.meshgrid(lats,lons)

lat_lon_pairs = np.vstack([lat_grid.ravel(), lon_grid.ravel()]).T

df_latlon = pd.DataFrame(lat_lon_pairs, columns = ['Latitude','Longitude'])

all_times = np.repeat(filtered_times_df.index, len(lat_lon_pairs))
repeated_lat_lon_pairs = np.tile(lat_lon_pairs, (len(filtered_times_df.index),1))

df_precip = pd.DataFrame(repeated_lat_lon_pairs, columns = ['Latitude','Longitude'])
df_precip['Time'] = all_times
 
pr_data = {}

total_time = 0
df_len = len(df_precip)
 
for index, row in df_precip.iterrows():
    iter_s_time = time.time()
    
    result = pr_monthly.identify(geometry={"x":row['Longitude'],"y":row['Latitude']},slice_id=row['Time'])
    pr_data[index] = result['value']
    
    iter_e_time = time.time()
    total_time = total_time + (iter_e_time - iter_s_time)
    average_time = total_time / (index + 1)
    
    # Status of for-loop

    clear_output(wait = True)

    #display('Longitude: ' + str(row['Longitude']) + ' Latitude: ' + str(row['Latitude']) + ' Time: ' + str(row['Time']))
    display('Iteration ' + str(index+1) + ' of ' + str(df_len))
    display('Average Time Per Iteration (seconds): ' + str(round(average_time,2)))
    display('Total Time Elasped (minutes): ' + str(round(total_time/60,2)))
    display('Estimated Time Remaining (minutes): ' + str(round((df_len-index)*average_time/60,2)))


df_precip['Precipitation (inches)'] = pr_data
 
display(df_precip)
```

Now that we have precipitation measurements for each locations and times, we need to update the slice values for the actual time values.

```python
df_precip['Time'] = df_precip['Time'].map(filtered_times_df['GMT Time'])
df_precip
```

Next we turn the precipitation dataframe into an xarray data array, which is great for handling three dimensional data arrays (e.g., latitude, longitude, and time). 

```python
import xarray as xr

ds = df_precip.set_index(['Latitude','Longitude','Time']).to_xarray()

#data_array = ds['Precipitation (inches)']

#data_array
```

Then we need to remove the hour, minutes, and seconds from the time values so that they only consist of the date. For simplicity, the day of each month is updated to the first of the month. However, you may not want to do that if you are conducting rigorous analyses.

```python
years = ds.Time.dt.year
months = ds.Time.dt.month

updated_times = [pd.Timestamp(year = y, month = m, day = 1) for y,m in zip(years.values, months.values)]

ds.coords['Time'] = ('Time', updated_times)

ds
```

The last thing to do before visualizing and working with the data is to replace the areas that have "NoData" values with "NaN" (not-a-number values) to allow these values to be plotted. Areas with "NoData" values are simply areas over oceans.

```python
ds_masked = ds.where(~ds.isin(['NoData']), np.nan)
ds_masked = ds_masked.astype(float)

ds_masked
```

The data is finally ready to be visualized!

First, let's look at the first time slice of the precipitation data, which is associated with July of 2015.

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

pr_var = ds_masked['Precipitation (inches)']
pr_slice = pr_var.sel(Time = '2015-07-01').drop('Time')

fig, ax = plt.subplots(subplot_kw = {'projection': ccrs.PlateCarree()})
ax.set_extent([-130,-60,22,50], crs = ccrs.PlateCarree())

contour = pr_slice.plot.contourf(ax = ax, transform = ccrs.PlateCarree(), cmap = 'RdBu', levels = 50)
ax.add_feature(cfeature.BORDERS, edgecolor = 'black')
ax.coastlines()

plt.show()
```

According to the data, this model run of the CanESM5 projected for the month of July 2015 that the Midwest was would receive quite a bit of rain compared to the West. 

The coarse resolution at which we sampled points is the reason for the coarse resolution of the plot. Sampling of more points would lead to finer resolution.

**Note:** Remember that Earth System Models (ESMs) are not meant to be interpreted outside of the scope of the parameters on which they were run. ESMs are usually interpreted with multiple model runs, different model parameters, and/or multiple ESMs to assess averages, trends, and patterns. So the data we see here is just one data point and not meant to be the authoritative source of precipitation data for that month and year over the US.

Now, let's look at the average over all the months of July from 2015 to 2100. 

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

ds_masked = ds.where(~ds.isin(['NoData']), np.nan)
ds_masked = ds_masked.astype(float)
ds_masked = ds_masked.mean(dim = 'Time')

pr_var = ds_masked['Precipitation (inches)']

fig, ax = plt.subplots(subplot_kw = {'projection': ccrs.PlateCarree()})
ax.set_extent([-130,-60,22,50], crs = ccrs.PlateCarree())

contour = pr_var.plot.contourf(ax = ax, transform = ccrs.PlateCarree(), cmap = 'RdBu', levels = 50)
ax.add_feature(cfeature.BORDERS, edgecolor = 'black')
ax.coastlines()

plt.show()
```

From this we can see that, in general, the East coast and Midwest are projected to receive more rain on average for the month of July compared to the West.

Last, let's take a look at how the decadal average for the month of July at the beginning of the century (2015-2024) compares to the decadal average at the end of the century (2091-2100) to see which areas are projected to get more or less precipitation.

```python
pr_2015_2024 = ds.sel(Time = slice('2015-07-01','2024-07-31'))
pr_2091_2100 = ds.sel(Time = slice('2091-07-01','2100-07-31'))

pr_mean_2015_2024 = pr_2015_2024.where(~pr_2015_2024.isin(['NoData']), np.nan)
pr_mean_2015_2024 = pr_mean_2015_2024.astype(float)
pr_mean_2015_2024 = pr_mean_2015_2024.mean(dim = 'Time')

pr_mean_2091_2100 = pr_2091_2100.where(~pr_2091_2100.isin(['NoData']), np.nan)
pr_mean_2091_2100 = pr_mean_2091_2100.astype(float)
pr_mean_2091_2100 = pr_mean_2091_2100.mean(dim = 'Time')

pr_var_2015_2024 = pr_mean_2015_2024['Precipitation (inches)']
pr_var_2091_2100 = pr_mean_2091_2100['Precipitation (inches)']
pr_var_dif = pr_var_2091_2100 - pr_var_2015_2024

fig, ax = plt.subplots(subplot_kw = {'projection': ccrs.PlateCarree()})
ax.set_extent([-130,-60,22,50], crs = ccrs.PlateCarree())

contour = pr_var_dif.plot.contourf(ax = ax, transform = ccrs.PlateCarree(), 
                                   cmap = 'RdBu', levels = 50,
                                   vmin = -0.5, vmax = 1.5)
ax.add_feature(cfeature.BORDERS, edgecolor = 'black')
ax.coastlines()

plt.show()
```

From the plot, we can see that the decadal average in the Plains and Texas later in the century is projected to receive more precipitation compared to the decadal average earlier in the century. In contrast, parts of the East coast and the West are projected to receive less precipitation.

**Note:** These are very simple analyses intended for this tutorial and not meant to be used for identifying trends for scientific publication or other research interpretations.

## Bonus Material - Making a GIF of Monthly Precipitation from 2015 to 2100

It's also possible to visualize all of the slices of monthly precipitation in a GIF, which can be very important for visualizing and detecting patterns of interest (and it's really fun).

To visualize the results of all of the slices, we first create a list of urls that only differ in a single number from the SliceID. We use a for loop to create these urls. Then we store each image from their respective URLs and save them into an animated GIF.

Note: The cell below may take a few minutes to run as it is storing and animating over 1000 months/slices/images of monthly precipitation data.

```python
import imageio

# Create the first and second parts of the URLs that are the same for all of them

main_url_first = 'https://gis.earthdata.nasa.gov/UAT/rest/services/NEX-GDDP-CMIP6/pr_monthly_CanESM5_ssp126/ImageServer/exportImage?bbox=-180.0%2C-60.0%2C180.0%2C90.0&bboxSR=&size=1200%2C900&imageSR=4326&time=&format=jpgpng&pixelType=F32&noData=&noDataInterpretation=esriNoDataMatchAny&interpolation=RSP_BilinearInterpolation&compression=&compressionQuality=&bandIds=&sliceId='
main_url_last = '&mosaicRule=&renderingRule=&adjustAspectRatio=true&validateExtent=false&lercVersion=1&compressionTolerance=&f=image'

# Create the URL list with each URL having a different number for it's respective slice (month)

urls_list = [f"{main_url_first}{i}{main_url_last}" for i in range(1033)]

# Get the images and store them in the "image" variable (this takes several minutes to run)

images = []
for url in urls_list:
    response = requests.get(url)
    image = imageio.imread(response.content)
    images.append(image)

# Use all of the images to create a GIF called "monthly_precip.gif" where each image is shown for 0.25 seconds
    
imageio.mimsave('monthly_precip.gif', images, duration = 0.25)
```
